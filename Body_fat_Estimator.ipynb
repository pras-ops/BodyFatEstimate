* The code defines a function called correlation that takes in two arguments: df and threshold. The purpose of the function is to identify and return the column names of the independent variables in df that are highly correlated with one another (having a correlation coefficient greater than threshold).

* Here is a step-by-step explanation of the code:

* Initialize two variables: colcor and cormat. colcor is a set that will store the names of the columns that are highly correlated. cormat is a dataframe that stores the correlation matrix of the independent variables in df.

* Use two nested for-loops to iterate over each cell in the correlation matrix cormat. The outer loop is for the rows, and the inner loop is for the columns.

* In the inner loop, use an if-statement to check if the absolute value of the correlation coefficient in the current cell is greater than the specified threshold.

* If the condition is true, get the name of the column at the current row index and add it to the set colcor.

* Return the set colcor after the loops are completed.

* Call the correlation function and store the result in the variable ans, using a threshold value of 0.85.





```python
def correlation(df,threshold):
    
    colcor = set()
    cormat = df.corr()
    
    for i in range(len(cormat)):
        
        for j in range(i):
            
            '''   
            for each cell get the value of that cell by .iloc[i][j],where i is the 
            row and j is the col if that abs(value) is greater than the threshold, 
            get the col_name and add it in set
            
            '''
            
            if abs(cormat.iloc[i][j])>threshold:
                colname = cormat.columns[i]
                colcor.add(colname)
                
    return colcor
    
ans = correlation(train, threshold=0.85)

ans
```

The VIF value for each feature measures how much the variance of the estimated regression coefficients is increased when including other predictors in the model. * path = reg.cost_complexity_pruning_path(X_train,y_train) computes the cost complexity pruning path for the trained model reg.

* The cost complexity pruning is a method for controlling the complexity of a machine learning model, by adding a penalty term to the loss function being optimized during training. This penalty term is controlled by a hyperparameter ccp_alpha, and its purpose is to discourage the model from having too many non-zero coefficients (i.e., to reduce overfitting).

* The cost_complexity_pruning_path method returns the path of ccp_alpha values that were used to prune the model, which can be useful for finding the optimal value of ccp_alpha to control the complexity of the model.

* The code line ccp_alpha = path.ccp_alphas extracts the values of ccp_alpha that were used in the cost complexity pruning path. This array of ccp_alpha values can be used to choose the optimal value for controlling the model's complexity.

```python
path = reg.cost_complexity_pruning_path(X_train,y_train)
ccp_alpha = path.ccp_alphas
```

### Hyperparameter Tuning
###### The code defines a dictionary called "params" that contains hyperparameters for three different machine learning models: RandomForest, Decision Tree (D-tree), and Support Vector Machine (SVM).

* 'n_estimators': A list of 10 integers from 1 to 1200, representing the number of trees in the random forest.
* 'criterion': A list containing two strings, "mse" and "mae", representing the criterion used to measure the quality of the split.
* 'max_depth': A list of 5 integers from 1 to 30, representing the maximum depth of the tree.
* 'min_samples_split': A list of 4 integers, 2, 5, 10, and 12, representing the minimum number of samples required to split an internal node.
* 'min_samples_leaf': A list of 4 integers, 2, 5, 10, and 12, representing the minimum number of samples required to be at a leaf node.
* 'max_features': A list containing two strings, "auto" and "sqrt", representing the number of features to consider when looking for the best split.
* 'ccp_alpha': A list of 7 numbers, [1, 2, 2.5, 3, 3.5, 4, 5], representing the complexity parameter used for Cost-Complexity Pruning.

###### The second key-value pair is for the D-tree model, where the 'model' key contains an instance of the DecisionTreeRegressor class, and the 'params' key contains a dictionary of hyperparameters for the model. The hyperparameters for D-tree include:

* 'criterion': A list containing two strings, "mse" and "mae", representing the criterion used to measure the quality of the split.
* 'splitter': A list containing two strings, "best" and "random", representing the strategy used to choose the split at each node.
* 'min_samples_split': A list of 5 integers, [1, 2, 4, 10, 12], representing the minimum number of samples required to split an internal node.
* 'min_samples_leaf': A list of 5 integers, [1, 2, 3, 10, 12], representing the minimum number of samples required to be at a leaf node.
* 'max_features': A list containing two strings, "auto" and "sqrt", representing the number of features to consider when looking for the best split.
* 'ccp_alpha': A list of 7 numbers, [1, 2, 2.5, 3, 3.5, 4, 5], representing the complexity parameter used for Cost-Complexity Pruning.

###### The third key-value pair is for the SVM model, where the 'model' key contains an instance of the SVR class, and the 'params' key contains a dictionary of hyperparameters for the model. The hyperparameters for SVM include:

* 'C': A list of 4 numbers, [0.25, 0.50, 0.75, 1.0], representing the regularization parameter.
* 'tol': A list of 6 numbers, [1e-10, 1e-5, 1e-4, 0.025, 0.50, 0.

```python
params = {
    'Randomforest':{
        'model': RandomForestRegressor(),
        'params':{
            'n_estimators': [int(X) for X in np.linspace(start=1,stop=1200,num=10)],
            'criterion' :["mse","mae"],
            'max_depth':[int(X) for X in np.linspace(start=1,stop=30,num=5)],
            'min_samples_split':[2,5,10,12],
            'min_samples_leaf':[2,5,10,12],
            'max_features':['auto','sqrt'],
            'ccp_alpha':[1,2,2.5,3,3.5,4,5]
        }
    },
    
    'D-tree':{
        'model':DecisionTreeRegressor(),
        'params':{
            'criterion' :["mse","mae"],
            'splitter':['best','random'],
            'min_samples_split':[1,2,4,10,12],
            'min_samples_leaf':[1,2,3,10,12],
            'max_features':['auto','sqrt'],
            'ccp_alpha':[1,2,2.5,3,3.5,4,5],
        }
    },
    'SVM':{
        'model':SVR(),
        'params':{
            'C':[0.25,0.50,0.75,1.0],
            'tol':[1e-10,1e-5,1e-4,0.025,0.50,0.75],
            'kernel':['linear','poly','rbf','sigmoid'],
            'max_iter':[int(X) for X in np.linspace(start=1,stop=250,num=10)],
        }
    }
}
```

###### The code is performing a hyperparameter tuning using RandomizedSearchCV method in a loop over a list of models stored in the "params" dictionary.

* Line 1: Initialize an empty list "scores".

* Line 2: Start a loop over the models stored in "params" dictionary with the key "modelname" and values "mp".

* Line 3: Fit the randomized search cross-validation object "clf" with the parameters from the "params" dictionary for the current iteration "mp". The model used in the current iteration is given by "mp['model']", and its hyperparameters are given by "mp['params']". The cross-validation is performed with 5 folds, the number of jobs for parallel processing is set to -1 (meaning all cores available), the number of iterations for searching the hyperparameters is set to 10, and the scoring metric is set to negative mean squared error.

* Line 4: Fit the model "clf" to the training data "X_train" and target "y_train".

* Line 5-8: Append a dictionary with the model name "modelname", the best score obtained during the cross-validation "clf.best_score_", and the best hyperparameters "clf.best_estimator_" to the "scores" list.

* Line 9: End of the loop.

###### This code is performing a hyperparameter tuning for multiple models stored in the "params" dictionary using randomized search cross-validation. The best hyperparameters and scores for each model are stored in the "scores" list and can be used to select the best performing model.

```python
score = []
for modelname,mp in params.items():
    clf = RandomizedSearchCV(mp['model'],param_distributions=mp['params'],
                            cv=5,n_jobs=-1,n_iter=10,scoring='neg_mean_squared_error')
    clf.fit(X_train,y_train)
    score.append({
        'model_name':modelname,
        'best_score':clf.best_score_,
        'best_estimator':clf.best_estimator_
    })
```
